```python
bronze_df = spark.table("workspace.default.financialsBronze")
```

```python
from pyspark.sql.functions import regexp_replace, col

# Remove $ and commas, then cast to float
silver_df = bronze_df.withColumn(
    "Amount",
    regexp_replace(col("Amount"), "[$,]", "").cast("double")
)
```

```python
from pyspark.sql.functions import to_date

# format date column
silver_df = silver_df.withColumn("Date", to_date(col("Date"), "MM/dd/yyyy"))
```

```python
display(silver_df)
```

```python
silver_df.write.mode("overwrite").saveAsTable("workspace.default.financialsSilver")
```
